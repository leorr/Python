## a) Encontre DFT de x1[n] e plote (usando a função stem) os gráficos de sua magnitude e de sua fase.

```python
from scipy.fft import fft
import matplotlib.pyplot as plt
import numpy as np
import math as m

x=range(0,50)
y=[]

for n in range(0,26):
    y.append(n * m.exp(-0.3 * n))
for n in range(26, 50):
    y.append(0)

xK = fft(x,50)

plt.title('Magnitude')
plt.stem(x,np.abs(xK), use_line_collection=True)

print('\n')

plt.title('Fase')
plt.stem(x,np.angle(xK), use_line_collection=True)
```

# Questão 3. Nas questões abaixo, para cada função, plote a magnitude e a fase de sua DFT (crie as funções no Python):

##a) Dente de Serra, figura 1.
image.png

```python
from scipy.fft import fft
import matplotlib.pyplot as plt
import numpy as np
import math as m

x=range(0,81)
y=[]

for n in range(0,41):
    y.append(n)
for n in range(41, 81):
    y.append(n-40)

xK = fft(x,81)

plt.title('Dente de serra')
plt.stem(x,y, use_line_collection=True)
plt.show()

print('\n')

plt.title('Magnitude')
plt.stem(x,np.abs(xK), use_line_collection=True)
plt.show()

print('\n')

plt.title('Fase')
plt.stem(x,np.angle(xK), use_line_collection=True)
plt.show()
```

##Use os comandos do Python fft e fftshift para, numericamente, avaliar e plotar a Transformada Discreta de Fourier de x[n] e os seguintes sinais subamostrados em 500 valores de Ω no intervalo −π ≤ Ω ≤ π.

##a) y[n] = x[2n]

```python
from scipy.fft import fft
import matplotlib.pyplot as plt
import numpy as np
import math as m

x=range(0,501)
y=[]

for n in range(0, 51):
    y.append(2 * np.exp((-0.1*n) * (-0.1*n) * 0.5 ))
for n in range(51,501):
    y.append(0)


xK = fft(x,501)

plt.title('y(n) = x(2n)')
plt.stem(x,y, use_line_collection=True)
plt.show()

print('\n')

plt.title('Magnitude')
plt.stem(x,np.abs(xK), use_line_collection=True)
plt.show()

print('\n')

plt.title('Fase')
plt.stem(x,np.angle(xK), use_line_collection=True)
plt.show()
```
